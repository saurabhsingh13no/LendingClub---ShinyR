```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv("loan2.csv")
df
```

```python
type(df['emp_length'])
```

```python
pattern=r'\d+'
emp_length2=[]
for value in df['emp_length']:
#     print (value)
    k=re.findall(pattern,value)
    if (k):
#         print (k[0])
        emp_length2.append(k[0])
    else:
        emp_length2.append(0)

len(emp_length2)
```

```python
df['emp_length']=emp_length2
```

```python
df.to_csv("loan3.csv")
```
